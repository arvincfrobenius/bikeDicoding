## Visualization & Explanatory Analysis

### Pertanyaan 1:
###Bagaimana pengaruh musim terhadap jumlah pengguna sepeda tahun 2011 dan 2012?

````python
# Mengelompokkan data dan menghitung total penggunaan sepeda berdasarkan musim dan tahun
total_cnt_by_season_year = dataDay_df.groupby(['yr', 'season'])[['casual', 'registered']].sum()

# Menyiapkan data untuk visualisasi
total_cnt_by_season_year = total_cnt_by_season_year.reset_index()

# Mengatur data agar setiap kategori penggunaan sepeda dapat dibuat menjadi grafik batang
total_cnt_melted = total_cnt_by_season_year.melt(id_vars=['yr', 'season'], value_vars=['casual', 'registered'],
                                                 var_name='category', value_name='total')

# Membuat plot
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Membuat grafik batang untuk setiap kategori penggunaan sepeda berdasarkan musim dan tahun
sns.barplot(data=total_cnt_melted, x='season', y='total', hue='yr', ci=None, palette='Set2')

# Menambahkan judul dan label
plt.title('Total Penggunaan Sepeda Berdasarkan Musim dan Tahun')
plt.xlabel('Musim')
plt.ylabel('Total Penggunaan Sepeda')
plt.legend(title='Tahun')

# Mengatur label sumbu X untuk nama musim
plt.xticks(ticks=[0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'])

# Menampilkan plot
plt.show()
````

### Pertanyaan 2:


##Bagaimana penggunaan sepeda berdasarkan hari kerja senin hingga sabtu?




````python
# Mengelompokkan data dan menghitung total penggunaan sepeda berdasarkan hari kerja
total_weekday_by_day = dataDay_df.groupby(['weekday'])[['casual', 'registered']].sum()

# Mengatur ukuran dan gaya plot
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Membuat grafik garis untuk setiap kategori penggunaan sepeda berdasarkan hari kerja
sns.lineplot(data=total_weekday_by_day, x='weekday', y='casual', marker='o', label='Casual')
sns.lineplot(data=total_weekday_by_day, x='weekday', y='registered', marker='o', label='Registered')

# Menambahkan judul dan label
plt.title('Total Penggunaan Sepeda casual dan registered Berdasarkan hari kerja')
plt.xlabel('Hari kerja')
plt.legend(title='Kategori')

# Menampilkan plot
plt.show()
````

**Insight:**
- Total pengguna sepeda berdasarkan tahun dan musim menghasilkan data bahwa pada tahun 2011 lebih banyak dari pada tahun 2012 dan urutan berdasarkan musim terbanyak pada musim fall dan kedua adalah summer
- Total pengguna berdasarkan jam kerja menghasilkan analisis bahwa pada pengguna register lebih banyak dibanding casual

## Analisis Lanjutan (Opsional)

bagaimana trend pengguna sepeda casual dan registered berdasarkan tahun, bulan, musim, kondisi cuaca, dan hari kerja

````python

# Mengelompokkan data dan menghitung total penggunaan sepeda berdasarkan tahun, musim, dan kondisi cuaca
total_cnt_by_season = dataDay_df.groupby(['yr', 'season', 'weathersit','weekday'])[['casual', 'registered']].sum()
total_cnt_by_season['total'] = total_cnt_by_season['casual'] + total_cnt_by_season['registered']
#total_cnt_by_season = total_cnt_by_season.sort_values(by='season', ascending=False)

# Memeriksa hasil grup untuk debugging
print(total_cnt_by_season)
````

## menampilkan dalam bentuk visualisasi

````python
# Mengelompokkan data dan menghitung total penggunaan sepeda berdasarkan tahun, musim, kondisi cuaca, dan hari kerja
total_cnt_by_season = dataDay_df.groupby(['yr', 'season', 'weathersit', 'weekday'])[['casual', 'registered']].sum()
total_cnt_by_season['total'] = total_cnt_by_season['casual'] + total_cnt_by_season['registered']
total_cnt_by_season = total_cnt_by_season.reset_index()

# Menyiapkan data dalam format yang cocok untuk plot dengan Seaborn
total_cnt_melted = total_cnt_by_season.melt(id_vars=['yr', 'season', 'weathersit', 'weekday'],
                                            value_vars=['casual', 'registered', 'total'],
                                            var_name='category', value_name='count')

# Membuat visualisasi menggunakan Seaborn FacetGrid untuk memisahkan data berdasarkan tahun dan kondisi cuaca
g = sns.FacetGrid(total_cnt_melted[total_cnt_melted['category'] == 'total'],
                  col='weathersit', row='yr', margin_titles=True, height=4)

# Menggunakan grafik batang untuk menampilkan total penggunaan sepeda berdasarkan musim dan hari kerja
g.map_dataframe(sns.barplot, x='season', y='count', hue='weekday', ci=None, palette='viridis')

# Menambahkan judul dan mengatur label
g.set_axis_labels("Musim", "Total Penggunaan Sepeda")
g.set_titles(col_template="Kondisi Cuaca {col_name}", row_template="Tahun {row_name}")
g.add_legend(title='Hari dalam Minggu')  # Hari dalam seminggu: 0 = Minggu, 6 = Sabtu

# Menyesuaikan tampilannya
plt.xticks([0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Total Penggunaan Sepeda Berdasarkan Tahun, Musim, Kondisi Cuaca, dan Hari dalam Minggu")

# Menampilkan plot
plt.show()
````
